# Matrix Factorization

```python
def top_metrics(test_set, pred_algo, top_k_users = 20):
    # Lets select the top 20 users with the most movies in the test set to create a second test for top 1 and top 5 accuracies
    # Dict to find the count how many reviews each user has in the test set
    top_users = {}
    for user_id, _, _ in test_set:
        if user_id not in top_users.keys():
            top_users[user_id] = 1
        else:
            top_users[user_id] += 1
    # sort this dict
    sorted_dict = dict(sorted(top_users.items(), key=lambda item: item[1], reverse=True))

    test_user_list = list(sorted_dict.keys())[0:top_k_users]

    top_test_set = []
    for u_id, movie_id, rating in test_set:
        if u_id in test_user_list:
            top_test_set.append((u_id, movie_id, rating))

    top_predictions = pred_algo.test(top_test_set)
    top_1_correct = 0
    top_5_correct = 0
    top_10_correct = 0
    for user in test_user_list:
        curr_preds = []
        for pred in top_predictions:
            if pred.uid == user:
                curr_preds.append((pred.iid, pred.r_ui, pred.est))
        # Sort Values
        sorted_test = sorted(curr_preds, key=lambda x: x[1], reverse=True)
        sorted_pred = sorted(curr_preds, key=lambda x: x[2], reverse=True)

        # Top_1 
        top_1_item = [sorted_test[0:1][0][0]]
        top_1_val = sorted_test[0:1][0][1]
        top_1_preds = [sorted_pred[0:1][0][0]]

        for pred in sorted_test:
            if pred[1] == top_1_val and pred[0] not in top_1_item:
                top_1_item.append(pred[0])

        for pred in top_1_preds:
            if pred in top_1_item:
                top_1_correct +=1

        # Top_5
        top_5_item = [pred[0] for pred in sorted_test[0:5]]
        top_5_val = sorted_test[4:5][0][1]
        top_5_preds = [pred[0] for pred in sorted_pred[0:5]]

        for pred in sorted_test:
            if pred[1] == top_5_val and pred[0] not in top_5_item:
                top_5_item.append(pred[0])
        
        for pred in top_5_preds:
            if pred in top_5_item:
                top_5_correct +=1

        # Top_10
        top_10_item = [pred[0] for pred in sorted_test[0:10]]
        top_10_val = sorted_test[9:10][0][1]
        top_10_preds = [pred[0] for pred in sorted_pred[0:10]]

        for pred in sorted_test:
            if pred[1] == top_10_val and pred[0] not in top_10_item:
                top_10_item.append(pred[0])
        
        for pred in top_10_preds:
            if pred in top_10_item:
                top_10_correct +=1
    return top_1_correct/20, top_5_correct/100, top_10_correct/200
```

```python
results = pd.DataFrame(columns=["Name", "RMSE", "Runtime", "Top_1", "Top_5", "Top_10"])
```

```python
for n_fact in range(50, 201, 50):

    # Set up the reader to have the min/max values, and then load in the values from the dataframe
    reader = Reader(rating_scale=(min(filter_df["rating"]), max(filter_df["rating"])))
    data = Dataset.load_from_df(filter_df[['userId', 'movieId', 'rating']], reader)

    # Split the data so we can get RMSE metrics
    train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)

    start_time = time.time()
    # Use the SVD to do the MF algorithm, and train it on the train set, and test on test set
    svd = SVD(n_factors = n_fact)
    svd.fit(train_set)
    test_predictions = svd.test(test_set)
    # test_predictions has the following format: List of Tuples where the tuples have r_ui and est where r_ui is the ground truth and est is the prediction given by the model

    # Used to get dim of SVD
    # user_matrix = svd.pu 
    # item_matrix = svd.qi
    #
    # print(user_matrix.shape)
    # print(item_matrix.shape)

    # Calculate RMSE on the test set
    print(f"RMSE for n_factors = {n_fact}")
    rmse = accuracy.rmse(test_predictions)
    end_time = time.time()

    top_acc = top_metrics(test_set, svd)
    
    results = pd.concat([results, pd.DataFrame({"Name": [f"SVD, num_factors = {n_fact}"], "RMSE": rmse, "Runtime": end_time - start_time,
                                                 "Top_1": top_acc[0], "Top_5": top_acc[1], "Top_10": top_acc[2]})], ignore_index=True)
```

# Collaborative Filtering

```python
# Set up the reader to have the min/max values, and then load in the values from the dataframe
reader = Reader(rating_scale=(min(filter_df["rating"]), max(filter_df["rating"])))
data = Dataset.load_from_df(filter_df[['userId', 'movieId', 'rating']], reader)

# Split the data so we can get RMSE metrics 
# NOTE: Use same random state to keep it the same
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)

start_time = time.time()
# We Use KNNBasic for item-item collaborative filtering and do the same steps as before
item_item = KNNBasic(sim_options={'user_based': False})
item_item.fit(train_set)
item_item_pred = item_item.test(test_set)
end_time = time.time()
rmse = accuracy.rmse(item_item_pred)
top_acc = top_metrics(test_set, item_item)
results = pd.concat([results, pd.DataFrame({"Name": [f"Item-Item_CF"], "RMSE": [rmse], "Runtime": [end_time - start_time],
                                            "Top_1": top_acc[0], "Top_5": top_acc[1], "Top_10": top_acc[2]})], ignore_index=True)

start_time = time.time()
# We Use KNNBasic for user-user collaborative filtering and do the same steps as before
user_user = KNNBasic(sim_options={'user_based': True})
user_user.fit(train_set)
user_user_pred = user_user.test(test_set)
end_time = time.time()
rmse = accuracy.rmse(user_user_pred)
top_acc = top_metrics(test_set, user_user)
results = pd.concat([results, pd.DataFrame({"Name": [f"User-User_CF"], "RMSE": [rmse], "Runtime": [end_time - start_time],
                                            "Top_1": top_acc[0], "Top_5": top_acc[1], "Top_10": top_acc[2]})], ignore_index=True)

start_time = time.time()
# Item-Item with cosine distance collaborative filtering
item_item_cos = KNNBasic(sim_options={'name': 'cosine', 'user_based': False})
item_item_cos.fit(train_set)
item_item_pred_cos = item_item_cos.test(test_set)
end_time = time.time()
rmse = accuracy.rmse(item_item_pred_cos)
top_acc = top_metrics(test_set, item_item_cos)
results = pd.concat([results, pd.DataFrame({"Name": [f"Item-Item_CF_Cos"], "RMSE": [rmse], "Runtime": [end_time - start_time],
                                            "Top_1": top_acc[0], "Top_5": top_acc[1], "Top_10": top_acc[2]})], ignore_index=True)

start_time = time.time()
# User-User with cosine distance collaborative filtering
user_user_cos = KNNBasic(sim_options={'name': 'cosine', 'user_based': True})
user_user_cos.fit(train_set)
user_user_pred_cos = user_user_cos.test(test_set)
end_time = time.time()
rmse = accuracy.rmse(user_user_pred_cos)
top_acc = top_metrics(test_set, user_user_cos)
results = pd.concat([results, pd.DataFrame({"Name": [f"User-User_CF_Cos"], "RMSE": [rmse], "Runtime": [end_time - start_time],
                                            "Top_1": top_acc[0], "Top_5": top_acc[1], "Top_10": top_acc[2]})], ignore_index=True)

print(f"Item-Item CF RMSE: {accuracy.rmse(item_item_pred)}")
print(f"User-User CF RMSE: {accuracy.rmse(user_user_pred)}")
print(f"Item-Item Cosine CF RMSE: {accuracy.rmse(item_item_pred_cos)}")
print(f"User-User Cosine CF RMSE: {accuracy.rmse(user_user_pred_cos)}")
```
